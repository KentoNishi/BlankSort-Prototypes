```python
for i in range(3):
    subdata=data[i]
    metrics=[item[:3] for item in subdata]
    X = ['BlankSort','MultipartiteRank','RAKE','TextRank','TopicRank']
    fig, ax = subplots(figsize=(8,5))
    df = pd.DataFrame(metrics, index=X)
    df.plot.bar(rot=0,ax=ax, width=0.5,title=datasets[i]+" Dataset - Performance")
    ax.legend(["Precision", "Recall","F1-Score"],loc="upper center")
    ax.yaxis.grid()
    ax.set_axisbelow(True)
    fig.show()
    fig.savefig("C:\\Users\\kento\\Documents\\GitHub\\BlankSort-Prototypes\\images\\charts\\"+datasets[i]+"Performance.svg",format="svg")
```

```python
for i in range(3):
    subdata=data[i]
    metrics=[item[3] for item in subdata]
    maxHeight=max(metrics)
    X = ['BlankSort','MultipartiteRank','RAKE','TextRank','TopicRank']
    fig, ax = subplots(figsize=(8,5))
    df = pd.DataFrame(metrics, index=X)
    df.plot.bar(rot=0,ax=ax, width=0.5,title=datasets[i]+" Dataset - Execution Time")
    ax.legend(["Time (ms)"],loc="upper center")
    ax.set_ylabel("Execution Time (milliseconds)")
    for p in ax.patches:
        ax.annotate(np.round(p.get_height(),decimals=2), (p.get_x()+p.get_width()/2., max(0,p.get_height()-maxHeight*0.02)), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
    ax.yaxis.grid()
    ax.set_axisbelow(True)
    fig.show()
    fig.savefig("C:\\Users\\kento\\Documents\\GitHub\\BlankSort-Prototypes\\images\\charts\\"+datasets[i]+"Time.svg",format="svg")
```
